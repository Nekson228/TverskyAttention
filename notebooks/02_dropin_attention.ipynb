```python
import torch
import torch.nn as nn

from src.arhitecture import TverskyMultiHeadAttentionDropIn
```

```python
batch_size = 2
seq_length = 16
d_model = 64

X = torch.randn(batch_size, seq_length, d_model)
```

```python
tversky_attention = TverskyMultiHeadAttentionDropIn(
    d_model=d_model,
    num_heads=4,
    num_features=32,
    model_type="contrast",
)

result = tversky_attention(X)
assert result.shape == (batch_size, seq_length, d_model)

result.backward(torch.ones_like(result))
```

```python
attention = nn.MultiheadAttention(d_model, num_heads=4, batch_first=True)
result = attention(X, X, X)[0]
assert result.shape == (batch_size, seq_length, d_model)
```

```python
basic_params = sum(p.numel() for p in attention.parameters())
tversky_params = sum(p.numel() for p in tversky_attention.parameters())
print(f"Параметров в MultiheadAttention: {basic_params}")
print(f"Параметров в TverskyMultiHeadAttentionDropIn: {tversky_params}")
print(f"Разница: {tversky_params - basic_params} ({(tversky_params - basic_params) / basic_params:.2%})")
```
